```python
#Q1 Write a program that fetches data from a specific URL and print it on screen.

import requests

def fetch_data(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  
        return response.text
    except requests.exceptions.RequestException as e:
        return f"An error occurred: {e}"

url = "https://docs.python.org/3/"
data = fetch_data(url)
print(data)
```

```python
#Q2 Write a Program that computes total size of all the files in your current directory folder.

import os

def get_total_size(directory):
    total_size = 0  

    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        
        if os.path.isfile(file_path):
            total_size += os.path.getsize(file_path)  

    return total_size

current_directory = os.getcwd()  
total_size = get_total_size(current_directory)
print(f"Total size of all files in the current directory: {total_size} bytes")
```

```python
#Q3 Write a program that reads a file line by line and each line read from the file is copied and to another file with line numbers specified at the beginning of the line.

def copy_with_line_numbers(input_file, output_file):
    try:
        with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
            line_number = 1
            for line in infile:
                
                outfile.write(f"{line_number}: {line}")
                line_number += 1
        print(f"Lines copied successfully from {input_file} to {output_file} with line numbers.")
    except FileNotFoundError:
        print(f"Error: The file {input_file} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

input_filename = 'file.txt'
output_filename = 'student.txt'
copy_with_line_numbers(input_filename, output_filename)
```

```python
#Q4 Write a program that counts the number of tabs, spaces and newline characters in a file.

def count_characters(file_path):
    try:
        with open(file_path, 'r') as file:
            tab_count = 0
            space_count = 0
            newline_count = 0
            
            for line in file:
                tab_count += line.count('\t')
                space_count += line.count(' ')
                newline_count += line.count('\n')
                
        return tab_count, space_count, newline_count
    except FileNotFoundError:
        return "Error: File not found."
    except Exception as e:
        return f"An error occurred: {e}"

file_path = 'student.txt'
tab_count, space_count, newline_count = count_characters(file_path)
print(f"Tabs: {tab_count}, Spaces: {space_count}, Newlines: {newline_count}")
```

```python
#Q5 Write a program that reads data from the file and calculates the percentage of vowels and consonants in the file.

def calculate_vowel_consonant_percentage(file_path):
    vowels = "AEIOUaeiou"
    total_chars = 0
    vowel_count = 0
    consonant_count = 0
    
    try:
        with open(file_path, 'r') as file:
            for line in file:
                for char in line:
                    if char.isalpha():  # Check if the character is a letter
                        total_chars += 1
                        if char in vowels:
                            vowel_count += 1
                        else:
                            consonant_count += 1
        
        vowel_percentage = (vowel_count / total_chars) * 100 if total_chars != 0 else 0
        consonant_percentage = (consonant_count / total_chars) * 100 if total_chars != 0 else 0
        
        return vowel_percentage, consonant_percentage
    
    except FileNotFoundError:
        return "Error: File not found."
    except Exception as e:
        return f"An error occurred: {e}"

file_path = 'file.txt'
vowel_percentage, consonant_percentage = calculate_vowel_consonant_percentage(file_path)
print(f"Vowels: {vowel_percentage:.2f}%")
print(f"Consonants: {consonant_percentage:.2f}%")
```

```python
#Q7 Write a program that writes data to a file in such a way that each character after a full stop is capitalized and all the numbers are written in brackets.

def process_text(text):
    new_text = ""
    capitalize_next = False

    for char in text:
        if char == '.':
            capitalize_next = True
            new_text += char
        elif capitalize_next and char.isalpha():
            new_text += char.upper()
            capitalize_next = False
        elif char.isdigit():
            new_text += f"({char})"
        else:
            new_text += char

    return new_text

def process_file(input_file, output_file):
    try:
        with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
            for line in infile:
                processed_line = process_text(line)
                outfile.write(processed_line)
        print(f"Data processed and written to {output_file}.")
    except FileNotFoundError:
        print(f"Error: The file {input_file} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

input_filename = 'file.txt'
output_filename = 'student.txt'
process_file(input_filename, output_filename)
```

```python
#Q8 Write a program that reads and copies the contents in another file. While copying replace all full stops with commas.

def replace_full_stops(input_file, output_file):
    try:
        with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
            for line in infile:
                modified_line = line.replace('.', ',')  # Replace full stops with commas
                outfile.write(modified_line)
        print(f"Contents copied successfully from {input_file} to {output_file} with full stops replaced by commas.")
    except FileNotFoundError:
        print(f"Error: The file {input_file} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

input_filename = 'file.txt'
output_filename = 'student.txt'
replace_full_stops(input_filename, output_filename)
```
